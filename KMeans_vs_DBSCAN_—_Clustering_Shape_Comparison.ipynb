```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs = pd.read_csv('/content/cluster_data.csv')
```

```python
blobs.head()
```

```python
sns.scatterplot(data=blobs, x='Feature 1', y='Feature 2')
```

```python
moons = pd.read_csv('/content/cluster_moons.csv')
```

```python
moons.head()
```

```python
sns.scatterplot(data=moons, x='X1', y='X2')
```

```python
circles = pd.read_csv('/content/circles.csv')
```

```python
circles.head()
```

```python
sns.scatterplot(data=circles, x='x', y='y')
```

```python
def display_categories(model,data):
  labels = model.fit_predict(data)
  sns.scatterplot(data=data, x='Feature 1', y='Feature 2', hue=labels, palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
display_categories(model,blobs)
```

```python
model = KMeans(n_clusters=2)
```

```python
def disp_categories(model,data):
  labels = model.fit_predict(data)
  sns.scatterplot(data=data, x='X1', y='X2', hue=labels, palette='Set1')
```

```python
disp_categories(model,moons)
```

```python
def display_category(model,data):
  labels = model.fit_predict(data)
  sns.scatterplot(data=data, x='x', y='y', hue=labels, palette='Set1')
```

```python
display_category(model,circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
```

```python
display_categories(model,blobs)
```

```python
model = DBSCAN(eps=0.15)
disp_categories(model,moons)
```

```python
model = DBSCAN(eps=0.15)
display_category(model,circles)
```
